```python
# Vérification des doublons
df_top_200.duplicated().sum()  
```

```python
# Vérification des valeurs manquantes
df_top_200.isnull().sum()  
```

```python
# Supprimer les lignes avec des valeurs nulles dans les colonnes artist et title 
df_top_200.dropna(subset=['artist', 'title'], inplace=True)
```

```python
# Transforer la colonne date en format date
df_top_200['date'] = pd.to_datetime(df_top_200['date'])
```

```python
# Remplacer les valeurs manquantes de la colonne streams par 0 
df_top_200['streams'].fillna('0', inplace=True)
```

```python
# Supprimer la colonne url 
df_top_200.drop('url', axis=1, inplace=True)
```

```python
df_top_200
```

```python
date_max = df_top_200['date'].max()
date_min = df_top_200['date'].min()
print("Date maximale :", date_max)
print("Date minimale :", date_min)
```

```python
# Afficher toutes les valeurs uniques de la colonne region 
tendances_uniques = df_top_200['region'].unique()
print(tendances_uniques)
```

```python
# Filtrer la colonne region sur : global et france uniquement 
df_top_200 = df_top_200[df_top_200['region'].isin(['Global', 'France'])]
```

```python
# Filtrer la colonne chart sur : top200 uniquement 
df_top_200 = df_top_200[df_top_200['chart'] == 'top200']
```

```python
# Supprimer les index
df_top_200.reset_index(drop=True, inplace=True)
```

```python
df_top_200
```

```python
# Afficher le top 5 des lignes les plus streamées
top_5_streams = df_top_200.nlargest(15, 'streams')
top_5_streams
```

```python
top_3_artists = df_top_200['artist'].value_counts().head(3)
top_3_artists
```
